```python
from googletrans import Translator ,LANGUAGES
```

```python
def every_Language(text):
    translator = Translator()
    for code, language in LANGUAGES.items():
        try:
            translation = translator.translate(text, src='en', dest=code)
            print(f"{text} translated to {language} : {translation.text}")
        except Exception as e:
            print(f"Could not translate to {language} ({code}): {e}")
```

```python
text = input("Enter a sentence in English: ")
every_Language(text)
```
